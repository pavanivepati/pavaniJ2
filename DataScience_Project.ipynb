```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/content/Supermaket_Sales_Dataset.csv")
print(df)
print(df.head())
```

```python
x=df.iloc[:,:4]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
X = df[["Total"]]
Y = df["Rating"]
```

```python
#Logistic Regression
mdl = LinearRegression()
mdl.fit(X, Y)
pred = mdl.predict([[66]])
print("Predicted value (LR): ",pred[0])
print("Accuracy (LR): ",mdl.score(X[:100], Y[:100])*100)


plt.scatter(X['Total'], Y, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```

```python
#RANDOM FOREST
from sklearn.ensemble import RandomForestRegressor
mdl = RandomForestRegressor(n_estimators=100,max_depth=6)
mdl.fit(X, Y)
pred = mdl.predict([[66]])
print("Predicted value (RFR): ",pred[0])
print("Accuracy (RFR): ",mdl.score(X[:100], Y[:100])*100)
plt.scatter(X['Total'], Y, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```

```python
#LINEAR REGRESSION
mdl = LinearRegression()
mdl.fit(X, Y)
pred = mdl.predict([[6.575]])
print("Predicted value (LR): ",pred[0])
print("Accuracy (LR): ",mdl.score(X[:100], Y[:100])*100)

plt.scatter(X['Total'], Y, color='g')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```

```python
#SVR
X = df[["Total"]]
Y = df["Rating"]
from sklearn.svm import SVR
mdl = SVR(kernel = 'rbf')
mdl.fit(X, Y)
pred = mdl.predict([[6.575]])
print("Predicted value (SVR): ",pred[0])
print("Accuracy (SVR): ",mdl.score(X[:100], Y[:100])*100)

plt.scatter(X['Total'], Y, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```

```python
#SVC
X = df[["Total"]]
Y_= df["Invoice ID"]
from sklearn.svm import SVC
mdl = SVC(kernel='poly')
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (SVC): ",pred[0])
print("Accuracy (SVR): ",mdl.score(X[:100], Y_[:100])*100)
plt.scatter(X['Total'], Y_, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```

```python
#RANDOM FOREST CLASSIFICATION
from sklearn.ensemble import RandomForestClassifier
mdl = RandomForestClassifier(criterion='entropy')
mdl.fit(X, Y_)
pred = mdl.predict([[6.575]])
print("Predicted value (RFC): ",pred[0])
print("Accuracy (RFC): ",mdl.score(X[:100], Y_[:100])*100)

plt.scatter(X['Total'], Y_, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```

```python
#RANDOM FOREST REGRESSION
from sklearn.ensemble import RandomForestRegressor
mdl = RandomForestRegressor(n_estimators=100,max_depth=6)
mdl.fit(X, Y)
pred = mdl.predict([[6.575]])
print("Predicted value (RFR): ",pred[0])
print("Accuracy (RFR): ",mdl.score(X[:100], Y[:100])*100)

plt.scatter(X['Total'], Y, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Ratings')
plt.show()
```

```python
#DECISION TREE REGRESSION
from sklearn.tree import DecisionTreeRegressor
mdl =  DecisionTreeRegressor(max_depth=3)
mdl.fit(X, Y)
pred = mdl.predict([[6.575]])
print("Predicted value (DTR): ",pred[0])
print("Accuracy (DTR): ",mdl.score(X[:100], Y[:100])*100)

plt.scatter(X['Total'], Y, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```

```python
#DECISION TREE CLASSIFICATION
X = df[["Total","Rating"]]
Y = df[["Invoice ID"]]
from sklearn.tree import DecisionTreeClassifier
mdl = DecisionTreeClassifier(max_leaf_nodes=3, random_state=1)
mdl.fit(X, Y)
pred = mdl.predict([[6.575,7.049]])
print("Predicted value (DTC): ",pred[0])
print("Accuracy (DTC): ",mdl.score(X[:100], Y[:100])*100)
plt.scatter(X['Total'], Y, color='b')
plt.plot(X['Total'], mdl.predict(X),color='black',linewidth=3)
plt.xlabel('Total')
plt.ylabel('Rating')
plt.show()
```
